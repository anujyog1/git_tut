### Exercise 1: Creating and Indexing a Pandas Series

Problem Statement:
You are given the marks of 5 students in a Math test:

Marks: 78, 85, 92, 67, 88

Student Names: 'Anita', 'Rahul', 'Zoya', 'Karan', 'Ishaan'

Tasks:

Create a pandas.Series with marks as values and student names as the index.

Retrieve the marks obtained by the student 'Zoya'.

Slice the series to retrieve marks of the first three students.



```python
import pandas as pd

# Step 1: Create the Series
marks = [78, 85, 92, 67, 88]
students = ['Anita', 'Rahul', 'Zoya', 'Karan', 'Ishaan']
marks_series = pd.Series(marks, index=students)

# Step 2: Retrieve Zoya's marks
zoya_marks = marks_series['Zoya']

# Step 3: Slice the first three students
first_three = marks_series[:3]

# Print results
print("Marks Series:")
print(marks_series)
print("\nZoya's Marks:", zoya_marks)
print("\nFirst Three Students:")
print(first_three)
```

### Exercise 2: Operating on DataFrame Columns

Tasks:

Create a pandas.DataFrame with the following data:

Products: 'Product A', 'Product B', 'Product C'

January: [1000, 1200, 900]

February: [1100, 1150, 950]

March: [1050, 1300, 1000]

Calculate the total sales per product across all months (row-wise sum) and store it in a new column called 'Total'.

Add another column 'Average' to show the average monthly sales for each product.

```python
import pandas as pd

# Step 1: Create the DataFrame
data = {
    'Jan': [1000, 1200, 900],
    'Feb': [1100, 1150, 950],
    'Mar': [1050, 1300, 1000]
}
products = ['Product A', 'Product B', 'Product C']

df = pd.DataFrame(data, index=products)

# Step 2: Calculate total sales per product
df['Total'] = df.sum(axis=1)

# Step 3: Calculate average monthly sales per product
df['Average'] = df.mean(axis=1)

# Display the DataFrame
print(df)
```

### Simple Line Plot
Problem: Plot a line graph showing temperature over 7 days.
         days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
         temperature = [30, 32, 31, 29, 28, 27, 26]



```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [30, 32, 31, 29, 28, 27, 26]

plt.plot(days, temperature, marker='o')
plt.title('Temperature Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

### 2. Simple Scatter Plot
Problem: Plot height vs weight of 10 students using a scatter plot.
height = [150, 160, 165, 155, 170, 175, 160, 180, 158, 168]
weight = [50, 55, 58, 53, 65, 70, 60, 75, 56, 63]


```python
height = [150, 160, 165, 155, 170, 175, 160, 180, 158, 168]
weight = [50, 55, 58, 53, 65, 70, 60, 75, 56, 63]

plt.scatter(height, weight, color='green')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

### 3. Pie Chart
Problem: Show the distribution of students in four streams.
       labels = ['Science', 'Commerce', 'Arts', 'Vocational']
       sizes = [40, 30, 20, 10]

```python
labels = ['Science', 'Commerce', 'Arts', 'Vocational']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Student Distribution by Stream')
plt.axis('equal')
plt.show()
```
